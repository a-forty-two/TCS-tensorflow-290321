```python
mycols = ['id', 'radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean']
data_plot = data.loc[:, mycols]
outputs = data.loc[:, 'diagnosis']
# Benign (Blue) and Malignant (Red)
logic  = lambda val: 'red' if val=='M' else 'blue'
diagnosis_encoded = outputs.map(logic)

plt.figure()
sm = pd.plotting.scatter_matrix(data_plot, c=diagnosis_encoded, figsize=(15,15))
```

```python
semidata = data.loc[:,mycols]
rawdata = semidata.set_index('id')
rawdata.head()
```

```python
# ML => 
# f(inputs) = w1*input1 + w2*i2 + w3*i3..... wn*in + bias 
# millions-> wx*input-> always be large?
# 0.00001 -> wx*inputs-> always be small?
# creates a disadvantage for all other variables which are small
# SOLUTION> bring them down in the same scale
# NORMALIZATION-> normal (90 degrees)-> pi to radians, 0 to pi
# 0 to 1 , -inf to +inf, -1 to 1, ..... many many types 

# Colors-> (0,0,0) to (255,255,255)
# grayscale -> (0) to (255)
# profit/loss -> -inf to +inf 
# relations/correlation/impact-> -1 to 1
# waves/sound -> -1 to 1 

# Z-Score = (input-mean)/ standard_dev
# z-score = (x- mu) / sigma 

stats=rawdata.describe().T
stats
```

```python
# divide the data between training and testing
# calculate mu and sigma only for training data
# if your assumptions are well placed, then mu and sigma should be fine for testing data also!



mu = stats['mean']
sigma = stats['std']
scaled_inputs = (rawdata - mu)/ sigma 
scaled_inputs.head()
```

 2 square =4 (larger than original)

 0.1 square = 0.01 (smaller than original)

 -0.1 square = 0.01 (larger than org)


 math is inconsistent (-inf to -1], (-1 to 0], [0,1), [1,inf) 

```python
scaled_inputs.to_csv('zscored.csv')
!ls
# save this to GCS 
```

```python
scaled_inputs.info()
```

```python
logic  = lambda val: 1 if val=='M' else 0
y = outputs.map(logic)
x = scaled_inputs
```

```python
# the correct way of Z-Scoring 
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(rawdata, y, test_size=0.2, random_state=42)
stats = xtrain.describe().T
mu = stats['mean']
sigma = stats['std']
xtrain_scaled = (xtrain - mu)/ sigma 
xtest_scaled = (xtest-mu)/sigma 
```

```python
xtrain_scaled.head()
```

```python
xtest_scaled.head()
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.svm import SVC, LinearSVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

#comparison factors/metrics


algos_all = {'SGD':SGDClassifier(), 'KNN': KNeighborsClassifier(), 'RF': RandomForestClassifier(),
             'RF100': RandomForestClassifier(n_estimators=100),'RF50': RandomForestClassifier(n_estimators=50),
             'Sherlock': GaussianNB(), 'SVM': SVC(), 'LinearSVM': LinearSVC()}
import time
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
import numpy as np
models = []
scores_all = []
best_model = 0.0
best_acc = 0.0
best_cvs = 0.0
for name, algo in algos_all.items():
  print('......initializing.... algo name: '+ name)
  starttime = time.time()
  model = algo
  model.fit(xtrain_scaled, ytrain)
  predictions = model.predict(xtest_scaled)
  # we already know the correct answers from ytest!
  # we will compare the predictions v/s the known answers to check if predictions were right or not
  endtime = time.time()
  accuracy_current = accuracy_score(predictions, ytest)
  cv_score = cross_val_score(model, x, y, cv=5)
  timetaken = endtime - starttime 
  cvscoremean = np.mean(cv_score)
  score_individual = (name, timetaken, accuracy_current, cvscoremean)
  models.append(model)
  if accuracy_current > best_acc:
    if np.mean(cv_score) > best_cvs:
      best_model= model
  scores_all.append(score_individual)
  print('Time taken: ' + str(timetaken))
  print('Accuracy: '+ str(accuracy_current))
  print('CV Scores: '+ str(cvscoremean))
```

```python
# constant through learning-> Placeholders [User provides as inputs-> parameters]
# values that change during learning-> Variables [weights, biases]

trainingInputs = tf.placeholder(shape=[None, 3], dtype=tf.float32)
trainingOutputs = tf.placeholder(shape=[None, 1], dtype=tf.float32)
weights = tf.Variable(initial_value=[[0.03],[0.1],[0.01]], dtype=tf.float32)
bias = tf.Variable(initial_value= [[0.005]], dtype=tf.float32)
```

```python
# build a graph out of nodes 
y = tf.matmul(trainingInputs, weights) + bias
predictions = tf.nn.sigmoid(y)
loss = tf.reduce_sum(trainingOutputs - predictions) # Mean Absolute Error 
# hyperparameters -> values that developers set and are NOT provided by users
# if users want to set HPs, they should provide via command line arguments 
HP_learningRate = 0.05
# this is used to reduce the losses in the next iteration
optimizer = tf.train.GradientDescentOptimizer(learning_rate=HP_learningRate).minimize(loss)
```
