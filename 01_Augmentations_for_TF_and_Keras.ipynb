```python
import Augmentor
pipes = Augmentor.Pipeline('images')
```

```python
pipes.flip_left_right(0.5)
pipes.black_and_white(0.25)
pipes.random_distortion(1, 10,10, 1)
pipes.skew(0.25, 0.5)
pipes.sample(10)
```
