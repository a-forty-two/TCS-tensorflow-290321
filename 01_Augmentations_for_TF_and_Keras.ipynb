```python
from keras.preprocessing.image import img_to_array, ImageDataGenerator, load_img
generator = ImageDataGenerator(rotation_range=40, shear_range=0.25, zoom_range=0.25,
                               horizontal_flip=True, brightness_range=(0.5, 1.5))
# brightness-> 50% to 150%
```

```python
img1 = load_img('images/2.jpg')
img2 = load_img('images/5.jpg')
i1 = img_to_array(img1)
i2 = img_to_array(img2)
```

```python
# before passing image into model-> make sure that input dimension == expected dimension on the network

i1.shape
# but-> NN such as VGG, RESNET, expect (:,:,:,3) as image!!! 
```

```python
sequenceimage1 = i1.reshape((1,) + i1.shape)
sequenceimage2 = i2.reshape((1,) + i2.shape)
```

```python
counter = 0
for batch in generator.flow(sequenceimage1, batch_size=1, save_to_dir='myaugs', save_format='jpeg',
                            save_prefix='pikachu'):
  if counter == 20:
    break
  counter = counter + 1

```
